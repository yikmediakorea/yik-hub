```python
import os
import re
import json
from googleapiclient.discovery import build
from langdetect import detect, DetectorFactory
import pandas as pd
from datetime import datetime, timedelta

DetectorFactory.seed = 0

# ============================================================
# 1. 설정
# ============================================================
API_KEY = "AIzaSyA2WZOzBO59dsdDaluGYLQ9_M_Uc3C3O3U"
youtube = build("youtube", "v3", developerKey=API_KEY)

# ============================================================
# 2. 채널 ID 추출 (URL에서 자동)
# ============================================================
def get_channel_id(channel_url):
    """YouTube 채널 URL에서 channel ID를 추출"""

    # @핸들 형식
    if "/@" in channel_url:
        handle = channel_url.split("/@")[1].split("/")[0].split("?")[0]
        response = youtube.search().list(
            part="snippet",
            q=handle,
            type="channel",
            maxResults=1
        ).execute()
        if response["items"]:
            return response["items"][0]["snippet"]["channelId"]

    # /channel/ 형식
    if "/channel/" in channel_url:
        return channel_url.split("/channel/")[1].split("/")[0].split("?")[0]

    # /c/ 또는 /user/ 형식
    for pattern in ["/c/", "/user/"]:
        if pattern in channel_url:
            username = channel_url.split(pattern)[1].split("/")[0].split("?")[0]
            response = youtube.search().list(
                part="snippet",
                q=username,
                type="channel",
                maxResults=1
            ).execute()
            if response["items"]:
                return response["items"][0]["snippet"]["channelId"]

    return None

# ============================================================
# 3. 채널 기본 정보 수집
# ============================================================
def get_channel_info(channel_id):
    """채널 구독자수, 총 조회수, 영상수 등 기본 정보"""
    response = youtube.channels().list(
        part="snippet,statistics,brandingSettings",
        id=channel_id
    ).execute()

    if not response["items"]:
        return None

    item = response["items"][0]
    return {
        "channel_id": channel_id,
        "channel_name": item["snippet"]["title"],
        "description": item["snippet"].get("description", ""),
        "subscriber_count": int(item["statistics"].get("subscriberCount", 0)),
        "total_views": int(item["statistics"].get("viewCount", 0)),
        "video_count": int(item["statistics"].get("videoCount", 0)),
        "country": item["snippet"].get("country", "N/A"),
        "created_at": item["snippet"]["publishedAt"],
    }

# ============================================================
# 4. 최근 영상 목록 수집 (롱폼/숏폼 구분)
# ============================================================
def get_recent_videos(channel_id, max_results=30):
    """최근 영상 목록과 조회수를 수집하고 롱폼/숏폼 구분"""

    # 업로드 플레이리스트 ID 가져오기
    channel_response = youtube.channels().list(
        part="contentDetails",
        id=channel_id
    ).execute()

    uploads_playlist_id = channel_response["items"][0]["contentDetails"]["relatedPlaylists"]["uploads"]

    # 최근 영상 ID 목록 가져오기
    video_ids = []
    next_page_token = None

    while len(video_ids) < max_results:
        playlist_response = youtube.playlistItems().list(
            part="contentDetails",
            playlistId=uploads_playlist_id,
            maxResults=min(50, max_results - len(video_ids)),
            pageToken=next_page_token
        ).execute()

        for item in playlist_response["items"]:
            video_ids.append(item["contentDetails"]["videoId"])

        next_page_token = playlist_response.get("nextPageToken")
        if not next_page_token:
            break

    # 각 영상의 상세 정보 가져오기
    videos = []
    for i in range(0, len(video_ids), 50):
        batch = video_ids[i:i+50]
        video_response = youtube.videos().list(
            part="snippet,statistics,contentDetails",
            id=",".join(batch)
        ).execute()

        for item in video_response["items"]:
            # 영상 길이 파싱 (ISO 8601)
            duration = item["contentDetails"]["duration"]
            seconds = parse_duration(duration)

            # 숏폼 판별 (60초 이하)
            is_short = seconds <= 60

            videos.append({
                "video_id": item["id"],
                "title": item["snippet"]["title"],
                "published_at": item["snippet"]["publishedAt"],
                "view_count": int(item["statistics"].get("viewCount", 0)),
                "like_count": int(item["statistics"].get("likeCount", 0)),
                "comment_count": int(item["statistics"].get("commentCount", 0)),
                "duration_seconds": seconds,
                "is_short": is_short,
                "language": detect_language_safe(item["snippet"]["title"]),
            })

    return videos

def parse_duration(duration_str):
    """ISO 8601 duration을 초로 변환 (PT1H2M3S → 3723)"""
    match = re.match(r'PT(?:(\d+)H)?(?:(\d+)M)?(?:(\d+)S)?', duration_str)
    if not match:
        return 0
    hours = int(match.group(1) or 0)
    minutes = int(match.group(2) or 0)
    seconds = int(match.group(3) or 0)
    return hours * 3600 + minutes * 60 + seconds

def detect_language_safe(text):
    """안전한 언어 감지"""
    try:
        return detect(text)
    except:
        return "unknown"

# ============================================================
# 5. 댓글 언어 분석 (시청자 국가 추정용)
# ============================================================
def analyze_comment_languages(video_id, max_comments=100):
    """영상 댓글의 언어 분포를 분석"""
    languages = {}

    try:
        response = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=min(100, max_comments),
            textFormat="plainText"
        ).execute()

        for item in response["items"]:
            comment_text = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            if len(comment_text.strip()) > 5:  # 너무 짧은 댓글 제외
                lang = detect_language_safe(comment_text)
                languages[lang] = languages.get(lang, 0) + 1
    except Exception as e:
        pass  # 댓글 비활성화 등

    # 비율로 변환
    total = sum(languages.values()) if languages else 1
    language_ratios = {
        lang: round(count / total * 100, 1)
        for lang, count in sorted(languages.items(), key=lambda x: -x[1])
    }

    return language_ratios

# ============================================================
# 6. 종합 분석 계산
# ============================================================
def analyze_channel(channel_url):
    """하나의 채널을 종합 분석"""

    print(f"\n{'='*60}")
    print(f"분석 중: {channel_url}")
    print(f"{'='*60}")

    # 채널 ID 추출
    channel_id = get_channel_id(channel_url)
    if not channel_id:
        print(f"  ❌ 채널 ID를 찾을 수 없습니다.")
        return None

    # 기본 정보
    info = get_channel_info(channel_id)
    if not info:
        print(f"  ❌ 채널 정보를 가져올 수 없습니다.")
        return None

    print(f"  채널명: {info['channel_name']}")
    print(f"  구독자: {info['subscriber_count']:,}")

    # 최근 영상 수집
    videos = get_recent_videos(channel_id, max_results=30)

    # 롱폼/숏폼 분리
    longform = [v for v in videos if not v["is_short"]]
    shortform = [v for v in videos if v["is_short"]]

    # 평균 조회수 계산
    longform_avg = (
        round(sum(v["view_count"] for v in longform) / len(longform))
        if longform else 0
    )
    shortform_avg = (
        round(sum(v["view_count"] for v in shortform) / len(shortform))
        if shortform else 0
    )

    # 조회율 (구독자 대비)
    engagement_rate = (
        round(longform_avg / info["subscriber_count"] * 100, 2)
        if info["subscriber_count"] > 0 else 0
    )

    # 댓글 언어 분석 (최근 롱폼 3개 영상에서)
    comment_languages = {}
    for video in longform[:3]:
        langs = analyze_comment_languages(video["video_id"], max_comments=50)
        for lang, ratio in langs.items():
            comment_languages[lang] = comment_languages.get(lang, 0) + ratio

    # 평균 내기
    if comment_languages:
        total = sum(comment_languages.values())
        comment_languages = {
            lang: round(val / total * 100, 1)
            for lang, val in sorted(comment_languages.items(), key=lambda x: -x[1])
        }

    # 콘텐츠 언어 추정 (영상 제목 언어 빈도)
    title_languages = {}
    for v in videos:
        lang = v["language"]
        title_languages[lang] = title_languages.get(lang, 0) + 1

    primary_language = max(title_languages, key=title_languages.get) if title_languages else "unknown"

    # 해외 비율 추정 (한국어가 아닌 댓글 비율)
    korean_ratio = comment_languages.get("ko", 0)
    overseas_ratio = round(100 - korean_ratio, 1)

    print(f"  롱폼 평균 조회수: {longform_avg:,}")
    print(f"  숏폼 평균 조회수: {shortform_avg:,}")
    print(f"  해외 시청자 추정: {overseas_ratio}%")
    print(f"  댓글 언어 분포: {comment_languages}")

    # 결과 딕셔너리
    result = {
        "채널 URL": channel_url,
        "채널명": info["channel_name"],
        "구독자 수": info["subscriber_count"],
        "총 조회수": info["total_views"],
        "총 영상 수": info["video_count"],
        "롱폼 영상 수 (최근 30)": len(longform),
        "숏폼 영상 수 (최근 30)": len(shortform),
        "롱폼 평균 조회수": longform_avg,
        "숏폼 평균 조회수": shortform_avg,
        "구독자 대비 조회율 (%)": engagement_rate,
        "콘텐츠 주 언어": primary_language,
        "댓글 언어 분포": json.dumps(comment_languages, ensure_ascii=False),
        "해외 시청자 비율 추정 (%)": overseas_ratio,
        "채널 설명": info["description"][:200],
        "분석 일시": datetime.now().strftime("%Y-%m-%d %H:%M"),
    }

    return result

# ============================================================
# 7. 전체 실행
# ============================================================
def run_batch_analysis(channel_urls, output_file="influencer_report.xlsx"):
    """여러 채널을 일괄 분석하고 엑셀로 출력"""

    results = []

    for url in channel_urls:
        try:
            result = analyze_channel(url)
            if result:
                results.append(result)
        except Exception as e:
            print(f"  ❌ 에러 발생: {e}")
            continue

    # 엑셀로 출력
    df = pd.DataFrame(results)
    df.to_excel(output_file, index=False, sheet_name="분석 결과")

    print(f"\n✅ 분석 완료! {len(results)}개 채널 → {output_file}")

    return df


# ============================================================
# 실행
# ============================================================
if __name__ == "__main__":

    # 분석할 채널 목록 (엑셀에서 가져온 것)
    channels = [
        "https://www.youtube.com/@ireneswnd",
        "https://www.youtube.com/@ChinguAmiga",
        "https://www.youtube.com/@KhoaiLangThang",
        "https://www.youtube.com/@HariJisun",
        "https://www.youtube.com/@cullen_hateberry",
        "https://www.youtube.com/@Doobydobap",
        "https://www.youtube.com/@Koreanosh",
        "https://www.youtube.com/@michellechoii",
        "https://www.youtube.com/@ryanbang",
        "https://www.youtube.com/@liryonni",
        "https://www.youtube.com/@KimbabFamily",
        "https://www.youtube.com/@biancakartika",
        "https://www.youtube.com/@kenhcuakem",
        "https://www.youtube.com/@thejamesrodriguezz",
        "https://www.youtube.com/YanReyzen",
        "https://www.youtube.com/@ggukim",
        "https://www.youtube.com/@SeoulMafia",
        "https://www.youtube.com/@IKITERUTV",
        "https://www.youtube.com/@mirasgarden",
        "https://www.youtube.com/@Q2HAN",
        "https://www.youtube.com/@klara_tania",
        "https://www.youtube.com/@MiLaOficial",
        "https://www.youtube.com/c/HannaCoreana",
    ]

    df = run_batch_analysis(channels)
    print(df.head())
```

```python
# ============================================================
# 🔍 인플루언서 즉석 분석기 - 링크만 넣으면 바로 분석!
# ============================================================

def instant_analyze():
    """유튜브 링크를 입력하면 즉시 분석 리포트 출력"""

    print("=" * 65)
    print("🌏 YIK Media 인플루언서 즉석 분석기")
    print("=" * 65)
    print()

    # 입력 받기
    url = input("📎 유튜브 채널 링크를 붙여넣으세요: ").strip()

    if not url:
        print("❌ 링크를 입력해주세요.")
        return

    print()
    print("⏳ 분석 중... (약 30초~1분 소요)")
    print("-" * 65)

    # 1. 채널 ID 추출
    channel_id = get_channel_id(url)
    if not channel_id:
        print("❌ 채널을 찾을 수 없습니다. URL을 확인해주세요.")
        return

    # 2. 기본 정보
    info = get_channel_info(channel_id)
    if not info:
        print("❌ 채널 정보를 가져올 수 없습니다.")
        return

    # 3. 최근 영상 수집
    videos = get_recent_videos(channel_id, max_results=30)

    # 4. 롱폼/숏폼 분리 및 계산
    longform = [v for v in videos if not v["is_short"]]
    shortform = [v for v in videos if v["is_short"]]

    longform_avg = round(sum(v["view_count"] for v in longform) / len(longform)) if longform else 0
    shortform_avg = round(sum(v["view_count"] for v in shortform) / len(shortform)) if shortform else 0

    engagement_rate = round(longform_avg / info["subscriber_count"] * 100, 2) if info["subscriber_count"] > 0 else 0

    # 5. 댓글 언어 분석
    comment_languages = {}
    analyzed_videos = 0
    for video in longform[:3]:
        try:
            langs = analyze_comment_languages(video["video_id"], max_comments=50)
            for lang, ratio in langs.items():
                comment_languages[lang] = comment_languages.get(lang, 0) + ratio
            analyzed_videos += 1
        except:
            continue

    if comment_languages:
        total = sum(comment_languages.values())
        comment_languages = {
            lang: round(val / total * 100, 1)
            for lang, val in sorted(comment_languages.items(), key=lambda x: -x[1])
        }

    korean_ratio = comment_languages.get("ko", 0)
    overseas_ratio = round(100 - korean_ratio, 1)

    # 6. 콘텐츠 언어
    title_languages = {}
    for v in videos:
        lang = v["language"]
        title_languages[lang] = title_languages.get(lang, 0) + 1
    primary_language = max(title_languages, key=title_languages.get) if title_languages else "unknown"

    # 언어 → 국가 매핑
    lang_to_country = {
        "id": "🇮🇩 인도네시아", "es": "🇲🇽 스페인어권(남미)",
        "vi": "🇻🇳 베트남", "th": "🇹🇭 태국",
        "ja": "🇯🇵 일본", "zh-cn": "🇨🇳 중국",
        "zh-tw": "🇹🇼 대만", "ko": "🇰🇷 한국",
        "en": "🇺🇸 영어권", "ru": "🇷🇺 러시아",
        "it": "🇮🇹 이탈리아", "fr": "🇫🇷 프랑스",
        "pt": "🇧🇷 포르투갈어권", "ar": "🇸🇦 아랍어권",
        "de": "🇩🇪 독일", "tl": "🇵🇭 필리핀",
    }

    primary_country = lang_to_country.get(primary_language, f"기타 ({primary_language})")

    # 구독자 단위 변환
    def format_subs(n):
        if n >= 10000000:
            return f"{n/10000:.0f}만"
        elif n >= 10000:
            return f"{n/10000:.1f}만"
        else:
            return f"{n:,}"

    def format_views(n):
        if n >= 10000:
            return f"{n/10000:.1f}만"
        else:
            return f"{n:,}"

    # 조건 충족 판단
    sub_ok = info["subscriber_count"] >= 500000
    view_ok = longform_avg >= 100000

    if sub_ok and view_ok:
        status = "✅ 조건 충족 (구독자 50만+ & 롱폼 평균 10만+)"
    elif sub_ok:
        status = "🔍 확인 필요 (구독자 OK, 롱폼 조회수 미달)"
    elif view_ok:
        status = "🔍 확인 필요 (조회수 OK, 구독자 미달)"
    else:
        status = "❌ 조건 미충족"

    # 최근 롱폼 조회수 추세
    if len(longform) >= 6:
        recent_3 = sum(v["view_count"] for v in longform[:3]) / 3
        older_3 = sum(v["view_count"] for v in longform[3:6]) / 3
        if recent_3 > older_3 * 1.1:
            trend = "📈 상승세"
        elif recent_3 < older_3 * 0.9:
            trend = "📉 하락세"
        else:
            trend = "➡️ 안정적"
    else:
        trend = "➡️ 데이터 부족"

    # ============================================================
    # 리포트 출력
    # ============================================================
    print()
    print("┌─────────────────────────────────────────────────────────────┐")
    print("│          📊 해외 타깃 글로벌 마케팅 인플루언서 분석 리포트       │")
    print("└─────────────────────────────────────────────────────────────┘")
    print()
    print(f"  채널명:          {info['channel_name']}")
    print(f"  채널 링크:       {url}")
    print(f"  카테고리:        (채널 설명 기반 추정 필요)")
    print()
    print(f"  ─── 기본 지표 ───")
    print(f"  구독자:          {format_subs(info['subscriber_count'])}명")
    print(f"  총 영상 수:      {info['video_count']}개")
    print(f"  총 조회수:       {format_views(info['total_views'])}회")
    print()
    print(f"  ─── 성과 지표 ───")
    print(f"  롱폼 평균 조회수:  {format_views(longform_avg)}회 (최근 {len(longform)}개 기준)")
    print(f"  숏폼 평균 조회수:  {format_views(shortform_avg)}회 (최근 {len(shortform)}개 기준)")
    print(f"  구독자 대비 조회율: {engagement_rate}%")
    print(f"  조회수 추세:      {trend}")
    print()
    print(f"  ─── 시청자 분석 ───")
    print(f"  콘텐츠 주 언어:    {primary_language} → {primary_country}")
    print(f"  해외 시청자 비율:  약 {overseas_ratio}% 추정")
    print(f"  댓글 언어 분포:")
    for lang, ratio in list(comment_languages.items())[:5]:
        country = lang_to_country.get(lang, lang)
        bar = "█" * int(ratio / 3) + "░" * (33 - int(ratio / 3))
        print(f"    {country:15} {bar} {ratio}%")
    print()
    print(f"  ─── 서울시 캠페인 적합성 ───")
    print(f"  선정 기준 충족:    {status}")
    print(f"  주요 타깃 권역:    {primary_country}")

    # 거주지 추정
    if korean_ratio > 50:
        residence = "국내 거주 추정 (한국어 댓글 다수)"
    else:
        residence = "해외 거주 추정 (항공·숙박 별도 필요 가능)"
    print(f"  거주지 추정:      {residence}")
    print()
    print(f"  ─── 최근 인기 영상 TOP 5 ───")
    top_videos = sorted(videos, key=lambda x: x["view_count"], reverse=True)[:5]
    for i, v in enumerate(top_videos, 1):
        vtype = "🎬" if not v["is_short"] else "📱"
        print(f"    {i}. {vtype} {v['title'][:45]}...")
        print(f"       조회수: {format_views(v['view_count'])}회")
    print()
    print(f"  ─── 채널 설명 ───")
    print(f"  {info['description'][:300]}")
    print()
    print("─" * 65)
    print(f"  📅 분석 일시: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
    print(f"  🏢 Powered by YIK Media Influencer Analyzer")
    print("─" * 65)

    # 다시 분석할지 물어보기
    print()
    again = input("🔄 다른 채널도 분석하시겠습니까? (y/n): ").strip().lower()
    if again == "y":
        instant_analyze()


# ============================================================
# 🚀 실행!
# ============================================================
instant_analyze()
```
